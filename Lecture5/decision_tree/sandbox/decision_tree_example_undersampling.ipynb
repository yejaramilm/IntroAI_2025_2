### import libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split
from sklearn import tree
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay, precision_score, recall_score, f1_score
from imblearn.over_sampling import SMOTE
```

### data prep

```python
df = pd.read_csv('raw/s41598-020-73558-3_sepsis_survival_primary_cohort.csv')
```

```python
plt.rcParams.update({'font.size': 16})

df.hist(figsize=(12,8))
```

#### Balance dataset by undersampling

```python
### balance dataset by undersampling positive class
def mkBalancedDataFrame(df):
    
    """
        FUNCTION TO CREATE DATASET THAT IS BALANCED WITH RESPECT TO HOSPITAL OUTCOME VARIABLE BY UNDERSAMPLING MAJORITY CLASS
        
        INPUT:
        - df = pandas DataFrame with 'hospital_outcome_1alive_0dead' column and 0 minority class as well as predictor variables
        
        OUTPUT:
        - df_balanced = pandas DataFrame with equal number of dead and alive outcomes
        
        DEPENDENCIES:
        - pandas 1.4.2
        - numpy 1.21.5
    """

    # create sample of dead outcomes
    df_sample_dead = df[df['hospital_outcome_1alive_0dead']==0]

    # get number of dead and alive outcomes
    num_samples_dead = len(df_sample_dead)
    num_samples_alive = df['hospital_outcome_1alive_0dead'].sum()

    # generate array of random integers of length num_samples_dead ranging between 0 and num_samples_alive
    arr_rand_int = np.random.randint(num_samples_alive, size=num_samples_dead)

    # use random int array to sample indexes of alive records from original df
    index_sample_alive = df[df['hospital_outcome_1alive_0dead']==1].index[arr_rand_int]

    # random sample of alive outcomes
    df_sample_alive = df.loc[index_sample_alive]

    # create balanced df
    df_balanced = pd.concat([df_sample_dead, df_sample_alive])
    
    return df_balanced
```

```python
# naive undersampling of majority class
df_balanced = mkBalancedDataFrame(df)
```

```python
# plot outcome histograms
plt.figure(figsize=(15,5))

plt.subplot(1,2,1)
df['hospital_outcome_1alive_0dead'].hist()
plt.title("Original")
plt.xlabel("Outcome")
plt.subplot(1,2,2)
df_balanced['hospital_outcome_1alive_0dead'].hist()
plt.title("Balanced")
plt.xlabel("Outcome")
```

#### train-test split

```python
# define predictor and target variable names
X_var_names = df.columns[:3]
y_var_name = df.columns[3]

# create predictor and target arrays
X = df_balanced[X_var_names]
y = df_balanced[y_var_name]

# create train and test datasets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
# define predictor and target variable names
X_var_names = df.columns[:3]
y_var_name = df.columns[3]

# create predictor and target arrays
X = df[X_var_names]
y = df[y_var_name]

X_resampled, y_resampled = SMOTE().fit_resample(X, y)

# create train and test datasets
X_train, X_test, y_train, y_test = train_test_split(X_resampled, y_resampled, test_size=0.2, random_state=0)
```

### Decision Tree

```python
# train model
clf = tree.DecisionTreeClassifier()
clf = clf.fit(X_train, y_train)
```

##### Display decision tree

```python
plt.figure(figsize=(24,16))

tree.plot_tree(clf)
plt.show()
```

##### Evaluate performance

```python
def evaluateModel(clf, X, y):

    # confusion matrix
    y_pred = clf.predict(X)
    cm = confusion_matrix(y, y_pred)
    cm_disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=['dead', 'alive'])
    cm_disp.plot()

    # print metrics
    print("Precision = " + str(np.round(precision_score(y, y_pred),3)))
    print("Recall = " + str(np.round(recall_score(y, y_pred),3)))
    print("F1 = " + str(np.round(f1_score(y, y_pred),3)))
```

```python
evaluateModel(clf, X_train, y_train)
```

```python
evaluateModel(clf, X_test, y_test)
```

#### Hyperparameter tuning

```python
# train model
clf_tuned = tree.DecisionTreeClassifier(max_depth=3) # set max depth to 3
clf_tuned = clf_tuned.fit(X_train, y_train)
```

##### Display decision tree

```python
plt.figure(figsize=(24,16))

tree.plot_tree(clf_tuned,feature_names=X_var_names, class_names=['dead', 'alive'])
plt.show()
```

##### Evaluate performance

```python
evaluateModel(clf_tuned, X_train, y_train)
```

```python
evaluateModel(clf_tuned, X_test, y_test)
```
